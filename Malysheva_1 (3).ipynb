—Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã, –∫–∞–∫–∏–µ –≤—ã–≤–æ–¥—ã –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å –æ –∑–Ω–∞—á–µ–Ω–∏—è—Ö?
–ü–æ—Å—Ç—Ä–æ–π—Ç–µ —Ç–æ—á–µ—á–Ω—É—é –¥–∏–∞–≥—Ä–∞–º–º—É —Ä–∞—Å—Å–µ–Ω–∏—è –ø–æ–ª—É—á–µ–Ω–Ω–æ–≥–æ —Ä—è–¥–∞

```python
import numpy as np
import statistics as st
import matplotlib.pyplot as plt

a = np.random.rand(5)
print("–ú–∞—Å—Å–∏–≤: ", a)
print("–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ: ", st.mean(a))
print("–ú–µ–¥–∏–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ: ", st.median(a))
plt.scatter([i for i in range(5)], a)
plt.show()
```

—Ñ—É–Ω–∫—Ü–∏–∏ (ùë•, –Ω–∞–ø—Ä–∏–º–µ—Ä, –∏–∑–º–µ–Ω—è–µ—Ç—Å—è –æ—Ç 1 –¥–æ 10). –í—ã–¥–µ–ª–∏—Ç—å —Å—Ä–µ–∑ –ø–µ—Ä–≤–æ–π
–ø–æ–ª–æ–≤–∏–Ω—ã –º–∞—Å—Å–∏–≤–∞ –∏ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –¥–ª—è –æ—Å–Ω–æ–≤–Ω–æ–≥–æ –º–∞—Å—Å–∏–≤–∞ ‚Äì
–ª–∏–Ω–µ–π–Ω—ã–π –∏ –¥–ª—è —Å—Ä–µ–∑–∞ ‚Äì —Ç–æ—á–µ—á–Ω—ã–π

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 11, 1)
y = np.sqrt(1 + np.e**np.sqrt(x) + np.cos(x**2))/np.abs(1 - np.sin(x)**3) + np.log(np.abs(2*x))
plt.plot(x, y)
plt.show()
plt.scatter(x[:6], y[:6])
plt.show()
```

–≤—ã—á–∏—Å–ª—è—é—â–µ–π –∏–Ω—Ç–µ–≥—Ä–∞–ª –ø–æ –ø—Ä–∞–≤–∏–ª—É —Ç—Ä–∞–ø–µ—Ü–∏–∏.

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import trapz

x = np.arange(1, 11, 1)
y = np.abs(np.cos(x*np.e**(np.cos(x) + np.log(x + 1))))
plt.plot(x, y)
plt.fill_between(x, y)
plt.show()
area = trapz(y)
print(area)
```

–°–æ–∑–¥–∞–π—Ç–µ –ø—Ä–æ—Å—Ç–µ–π—à–∏–π –∫–∞–ª—å–∫—É–ª—è—Ç–æ—Ä, –≤–∫–ª—é—á–∞—é—â–∏–π –æ—Å–Ω–æ–≤–Ω—ã–µ –¥–µ–π—Å—Ç–≤–∏—è
–¥–ª—è –¥–≤—É—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö ‚Ä≤ + ‚Ä≤, ‚Ä≤ ‚àí ‚Ä≤, ‚Ä≤ ‚àô ‚Ä≤, ‚Ä≤ √∑ ‚Ä≤.

```python
import numpy as np

#–ó–Ω–∞–∫ (+, - ,* ,/) –∏–ª–∏ —Ñ—É–Ω–∫—Ü–∏—è (1 - e**(x+y), 2 - cos(x+y), 3 - sin(x+y), 4 - x**y)
x = 3
y = 4
s = '4'
if s in ('+', '-', '*', '/', '1', '2', '3', '4'):
    if s == '+':
        print("%.2f" % (x+y))
    elif s == '-':
        print("%.2f" % (x-y))
    elif s == '*':
        print("%.2f" % (x*y))
    elif s == '/':
        if y != 0:
            print("%.2f" % (x/y))
        else:
            print("–î–µ–ª–µ–Ω–∏–µ –Ω–∞ –Ω–æ–ª—å!")
    elif s == '1':
        print("%.2f" % np.e**(x+y))
    elif s == '2':
        print("%.2f" % np.cos(x+y))
    elif s == '3':
        print("%.2f" % np.sin(x+y))
    elif s == '4':
        print("%.2f" % x**y)
else:
        print("–ù–µ–≤–µ—Ä–Ω—ã–π –∑–Ω–∞–∫ –æ–ø–µ—Ä–∞—Ü–∏–∏ –∏–ª–∏ —Ñ—É–Ω–∫—Ü–∏–∏!")
```
