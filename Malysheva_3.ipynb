Задание 1.3.1

Задайте 4 точки в трехмерном пространстве, рассчитайте между ними
расстояния по описанным в примере выше метрикам. Отобразите точки
в трехмерном пространстве.


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def printdistance(x, y):
  print("Для", x, "и", y)
  print("Расстояние Евклида", np.linalg.norm(np.array(x) - np.array(y)))
  print("Квадрат Евклидова расстояния", np.linalg.norm(np.array(x) - np.array(y)) ** 2)
  print("Расстояние Чебышева", np.linalg.norm(np.array(x) - np.array(y), ord=np.inf))
  print("Расстояние Хемминга", np.linalg.norm(np.array(x) - np.array(y), ord=1))
  fig = plt.figure()
  ax = fig.add_subplot(111, projection = '3d')
  ax.scatter(*zip(x))
  ax.scatter(*zip(y))
  plt.show()
  print("\n")

x = np.array([0, 1, 1])
y = np.array([3, 2, 1])
z = np.array([0, 1, 4])
w = np.array([0, 2, 1])

printdistance(x, y)
printdistance(x, z)
printdistance(x, w)
printdistance(y, z)
printdistance(y, w)
printdistance(z, w)
```
